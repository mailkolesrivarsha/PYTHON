SERIES

```python
import numpy as np
import pandas as pd
S=pd.Series([1,2,3,4,5,6,7])
print(S)
```

DESCRIPTIVE STATISTICS

```python
print(S.describe())
```

```python
print(S.mean())
print(S.std())
print(S.min())
print(S.max())
```

DATA MANIPULATION

```python
S=pd.Series([1,2,3,4,5,6,7])
mapping=S.map(lambda x:x*2)
print(mapping)
```

```python
applies=S.apply(lambda x:x**2)
print(applies)
```

```python
sorted=S.sort_values()
print(sorted)
```

```python
dropp=S.drop(2)
print(dropp)
```

HANDLING MISSING DATA

```python
S1=pd.Series([1,2,3,np.nan,7,0])
print(S1.isnull())
```

```python
print(S1.notnull())
```

```python
print(S1.fillna(0))
```

```python
print(S1.dropna())
```

INDEXING,SLICING,FILTERING

```python
S=pd.Series([1,2,3,4,5,6,7])
print(S.iloc[2])
print(S.iloc[-2])
print(S.loc[2])
greater_num=S[S>5]
print(greater_num)
```

AGGREAGATION

```python
print(S.sum())
print(S.cumsum())
print(S.aggregate(['sum','std','mean']))
```
