**Apply feature engineering techniques on a dataset to prepare it for machine learning.**

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
data = {
    'Age': [22, 35, np.nan, 28, 50, 41, 30, 60],
    'Salary': [25000, 50000, 45000, np.nan, 90000, 65000, 55000, 120000],
    'Gender': ['Male', 'Female', 'Female', np.nan, 'Male', 'Female', 'Male', 'Female'],
    'City': ['Hyderabad', 'Delhi', 'Mumbai', 'Hyderabad', np.nan, 'Delhi', 'Mumbai', 'Hyderabad'],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-20', '2024-04-18',
             '2024-05-12', '2024-06-25', '2024-07-30', '2024-08-05'],
    'Purchase_Amount': [5000, 8000, 7000, 6000, 12000, 9000, 7500, 15000],
    'Experience': [1, 5, 3, 2, 15, 10, 6, 20],
    'Rating': [4.1, 4.5, 3.9, 4.0, 4.8, 4.6, 4.2, 4.9]
}

df = pd.DataFrame(data)

print("="*60)
print("ORIGINAL DATASET")
print("="*60)
print(df)
```

Problem 1: Handling Missing Values

```python
print("\n")
print("="*60)
print("PROBLEM 1 - HANDLING MISSING VALUES")
print("="*60)

num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

print(df)
```

Problem 2: Feature Scaling

```python
print("\n")
print("="*60)
print("PROBLEM 2 - FEATURE SCALING")
print("="*60)

print("Before Scaling")
print(df[['Age','Salary']])

std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

df['Salary_Standardized'] = std_scaler.fit_transform(df[['Salary']])
df['Age_MinMax'] = minmax_scaler.fit_transform(df[['Age']])

print("\nAfter Scaling")
print(df[['Age','Age_MinMax','Salary','Salary_Standardized']])
```

Problem 3: Encoding

```python
print("\n")
print("="*60)
print("PROBLEM 3 - ENCODING")
print("="*60)

label = LabelEncoder()

df['Gender_Label'] = label.fit_transform(df['Gender'])

city_encoded = pd.get_dummies(df['City'], prefix='City')

df = pd.concat([df, city_encoded], axis=1)

print(df)
```

Problem 4: Feature Creation

```python
print("\n")
print("="*60)
print("PROBLEM 4 - FEATURE CREATION")
print("="*60)

df['Total_Spending'] = df['Salary'] + df['Purchase_Amount']

print(df[['Salary','Purchase_Amount','Total_Spending']])
```

Problem 5: Date Feature Extraction

```python
print("\n")
print("="*60)
print("PROBLEM 5 - DATE FEATURE EXTRACTION")
print("="*60)

df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()

print(df[['Date','Year','Month','Day','Weekday']])
```

Problem 6: Binning

```python
print("\n")
print("="*60)
print("PROBLEM 6 - BINNING")
print("="*60)

bins = [0,25,45,100]
labels = ['Young','Adult','Senior']

df['Age_Group'] = pd.cut(df['Age'],
                         bins=bins,
                         labels=labels)

print(df[['Age','Age_Group']])
```

Problem 7: Outlier Detection (IQR)

```python
print("\n")
print("="*60)
print("PROBLEM 7 - OUTLIER DETECTION")
print("="*60)

Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df['Salary'] < lower) | (df['Salary'] > upper)]

print("Outlier Rows:")
print(outliers)
```

Problem 8: Feature Interaction

```python
print("\n")
print("="*60)
print("PROBLEM 8 - FEATURE INTERACTION")
print("="*60)

df['Experience_Salary'] = df['Experience'] * df['Salary']

print(df[['Experience','Salary','Experience_Salary']])

print("\nImportance:")
print("Experience_Salary combines experience and salary,")
print("which may improve machine learning model performance.")
```

Problem 9: Feature Selection

```python
print("\n")
print("="*60)
print("PROBLEM 9 - FEATURE SELECTION")
print("="*60)

# Convert Age_Group to numeric for correlation
df['Age_Group_Code'] = LabelEncoder().fit_transform(df['Age_Group'])

numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

print("Correlation with Purchase_Amount")
print(corr['Purchase_Amount'].sort_values(ascending=False))

least_feature = corr['Purchase_Amount'].drop('Purchase_Amount').abs().idxmin()

print("\nLeast Important Feature:", least_feature)

final_df = df.drop(columns=[least_feature])

print("\nFinal Dataset")
print(final_df)

print("\n")
print("="*60)
print("FEATURE ENGINEERING COMPLETED SUCCESSFULLY")
print("="*60)
```
